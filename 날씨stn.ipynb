```python
def scatter_rainfall(df, stn4contest_value, ef_year_value, ef_month_value, ef_day_value):
    """
    주어진 조건에 따라 시간별 누적 강수량 그래프를 그리는 함수

    :param df: 데이터프레임
    :param stn4contest_value: rainfall_train.stn4contest 값
    :param ef_year_value: rainfall_train.ef_year 값
    :param ef_month_value: rainfall_train.ef_month 값
    :param ef_day_value: rainfall_train.ef_day 값
    """
        # 데이터 필터링
    filtered_df = df[(df['rainfall_train.stn4contest'] == stn4contest_value) &
                     (df['rainfall_train.ef_year'] == ef_year_value) &
                     (df['rainfall_train.ef_month'] == ef_month_value) &
                     (df['rainfall_train.ef_day'] == ef_day_value)]
    filtered_df = filtered_df.sort_values(by='rainfall_train.ef_year')
        # 그래프 그리는 거
    plt.figure(figsize=(10, 6))
    plt.scatter(filtered_df['rainfall_train.ef_hour'], filtered_df['rainfall_train.vv'], marker='o')
    plt.xlabel('Hour')
    plt.ylabel('Accumulated Rainfall (mm)')
    plt.title(f'Accumulated Rainfall for {stn4contest_value} on {ef_year_value} {ef_month_value}/{ef_day_value}')
    plt.xticks(filtered_df['rainfall_train.ef_hour'])  # X축에 시간 값 표시
    # plt.grid(True)
    plt.show()
```

```python
# 사용자 입력 받기
stn4contest_value = input('Enter stn4contest value: ')
ef_year_value = input('Enter ef_year value: ')
ef_month_value = int(input('Enter ef_month value (5, 6, 7, 8, 9): '))
ef_day_value = int(input('Enter ef_day value: '))

# 그래프 그리기 함수 호출
scatter_rainfall(df, stn4contest_value, ef_year_value, ef_month_value, ef_day_value)
```
